```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_task1 = pd.read_csv('./T1_train_loss.csv')
results_task2 = pd.read_csv('./T2_train_loss.csv')
results_task3 = pd.read_csv('./T3_train_loss.csv')
results_task4 = pd.read_csv('./Task4_train_loss.csv')
results_task5 = pd.read_csv('./Task5_train_loss.csv')
```

```python
val_results_task1 = pd.read_csv('./T1_val_accuracy.csv')
val_results_task2 = pd.read_csv('./T2_val_accuracy.csv')
val_results_task3 = pd.read_csv('./T3_val_accuracy.csv')
val_results_task4 = pd.read_csv('./T4_val_accuracy.csv')
val_results_task5 = pd.read_csv('./T5_val_accuracy.csv')
```

```python
X = range(len(list(results_task5.Step.values[:])))
Y1 = results_task1['good-thunder-1 - task_1_train_loss']
Y2 = results_task2['good-thunder-1 - task_2_train_loss']
Y3 = results_task3['good-thunder-1 - task_3_train_loss']
Y4 = results_task4['good-thunder-1 - task_4_train_loss']
Y5 = results_task5['good-thunder-1 - task_5_train_loss']
V1 = val_results_task1['good-thunder-1 - task_1_val_accuracy']
V2 = val_results_task2['good-thunder-1 - task_2_val_accuracy']
V3 = val_results_task3['good-thunder-1 - task_3_val_accuracy']
V4 = val_results_task4['good-thunder-1 - task_4_val_accuracy']
V5 = val_results_task5['good-thunder-1 - task_5_val_accuracy']
# plt.figure(nrows=1, ncols=2, figsize=(4,3), dpi=200)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3), dpi=150)
axes[0].plot(X,Y1)
axes[0].plot(X,Y2)
axes[0].plot(X,Y3)
axes[0].plot(X,Y4)
axes[0].plot(X,Y5)
axes[0].set_xlabel("Epochs")
axes[0].set_xticks(np.arange(min(X), max(X)+1, 5))
axes[0].set_ylabel("Train Loss")
axes[0].tick_params(axis='both', which='major', labelsize=8)
# ax.tick_params(axis='both', which='minor', labelsize=8)
axes[0].legend(['Task-1', 'Task-2', 'Task-3', 'Task-4', 'Task-5'],fontsize=8)
axes[1].plot(X,V1)
axes[1].plot(X,V2)
axes[1].plot(X,V3)
axes[1].plot(X,V4)
axes[1].plot(X,V5)
axes[1].set_xlabel("Epochs")
axes[1].set_xticks(np.arange(min(X), max(X)+1, 5))
axes[1].tick_params(axis='both', which='major', labelsize=8)
axes[1].set_ylabel("Valid Accuracy")
axes[1].legend(['Task-1', 'Task-2', 'Task-3', 'Task-4', 'Task-5'], fontsize=8)
plt.tight_layout()
plt.show()
plt.savefig('./hw2_compare_trainloss_val_acc.pdf', dpi=200)
```

```python
val_acc_orig = [70.28, 66.97, 72.67, 67.32, 69.31]
val_acc_comb = [15.1, 20.09, 6.5, 14.81, 12.13]
num_tasks = 5

plt.figure(figsize=(4,3), dpi=200)
X = np.arange(num_tasks)
# for i in range(num_tasks):
plt.bar(X, val_acc_orig, align='edge', width=0.2)
plt.bar(X, val_acc_comb, align='edge', width=-0.2)
plt.xticks(X, ['Task-Model 1', 'Task Model 2', 'Task Model 3', 'Task Model 4', 'Task Model 5'], fontsize=3.5)
plt.xlabel("Tasks")
plt.ylabel("Validation Accuracy")
plt.legend(['Task Val Accuracy', 'Combined Val Accuracy'], loc="upper left", fontsize=4)
plt.show()
plt.savefig('./hw2_compare_valAcc_combinedAcc.pdf', dpi=200)
```

```python
ensemble_acc = [13.05, 42.54, 43.12, 62.2, 68.66]
random_ensemble_acc = [18.05, 47.88, 48.2, 65.0, 66.78]
superclass_ensemble_acc = [22.67, 51.18, 50.77, 66.53, 64.39]

ensemble_name = ['Random', 'Confidence', 'Entropy', 'Stacked', 'Oracle']

plt.figure(figsize=(4,3), dpi=200)
X = np.arange(5)
# for i in range(num_tasks):
plt.bar(X, ensemble_acc, align='edge', width=-0.1, color='lightseagreen')
plt.bar(X, random_ensemble_acc, align='center', width=0.1, color='orange')
plt.bar(X, superclass_ensemble_acc, align='edge', width=0.1, color='blue')

plt.xticks(X, ensemble_name, fontsize=7)
plt.xlabel("Aggregation Methods")
plt.ylabel("Combined Val Accuracy")
plt.legend(['Original Split', 'Random Split', 'Superclass-based Split'], loc="upper left", fontsize=7)
plt.show()
plt.savefig('./hw2_compare_aggregation_combinedAcc.pdf', dpi=200)
```
