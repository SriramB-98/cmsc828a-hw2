```python
wordnet_is_a_txt_path = './wordnet.is_a.txt'
words_txt_path = 'words.txt'

with open(wordnet_is_a_txt_path, 'r') as f:
    wn_lines = f.readlines()
    
with open(words_txt_path, 'r') as f:
    w_lines = f.readlines()
    
child_to_parent_wnid = {}
for wn_line in wn_lines:
    parent_wnid, child_wnid = wn_line.split()
    child_to_parent_wnid[child_wnid] = parent_wnid
wnid_to_name: Dict[str, str] = {}
    
for w_line in w_lines:
    wnid, name = w_line.split('\t')
    wnid_to_name[wnid] = name

    
parent_wnid_set = set()
for child_wnid in wnids:
    if child_wnid=='n08742578':
        continue
    parent_wnid = child_to_parent_wnid[child_wnid]
    parent_wnid_set.add(parent_wnid)
parent_wnid_list = list(parent_wnid_set)
parent_wnid_list.sort()
print('the number of parent classes', len(parent_wnid_list))
```

```python
parent_idx_list: List[str] = []
parent_name_list: List[str] = []
for child_wnid in wnids:
    if child_wnid=='n08742578':
        parent_idx_list.append('None')
        parent_name_list.append('None')
        continue
    parent_wnid = child_to_parent_wnid[child_wnid]
    parent_name = wnid_to_name[parent_wnid].rstrip('\n')
    parent_idx = parent_wnid_list.index(parent_wnid)
    parent_idx_str = str(parent_idx)
    parent_idx_list.append(parent_idx_str)
    parent_name_list.append(parent_name)
assert len(parent_idx_list) == 200 and len(parent_name_list) == 200
# parent_idx_list_txt = '\n'.join(parent_idx_list)
# parent_name_list_txt = '\n'.join(parent_name_list)
```

```python
import random
task_to_class_list = {"1":[], "2":[], "3":[], "4":[], "5":[]}
tasks = [1, 2, 3, 4, 5]

class_ids_covered = []

for class_id in range(200):
    
    if class_id in class_ids_covered:
        continue;
        
    if parent_idx_list[class_id]=='None':
        siblings = [str(class_id)]
    else:
        curr_parent = parent_idx_list[class_id]
        siblings = []
        for i,x in enumerate(parent_idx_list):
            if (x == curr_parent):# and (i!=class_id):
                siblings.append(str(i))

#     print(class_id, siblings)
    class_ids_covered.extend(siblings)
#     print(class_ids_covered)
    class_ids_covered = list(set(class_ids_covered))
    siblings_task_assignment = random.sample(tasks,len(siblings))    
    
    for i,T in enumerate(siblings_task_assignment):
        task_to_class_list[str(T)].append(int(siblings[i]))
    
    for sib in siblings:
        for T in tasks:
            r = random.random()
            if r < 0.1:
                task_to_class_list[str(T)].append(int(sib))



for key_ in task_to_class_list.keys():
    task_to_class_list[key_] = list(set(task_to_class_list[key_]))
    print('Task:', key_)
    print(task_to_class_list[key_])
    print('number of classes:', len(task_to_class_list[key_]) )
    #         
    
```
